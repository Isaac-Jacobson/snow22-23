```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
img1 = Image.open('Flake000936_Cam1_1_2022-2-13-20-0-36-227.png')
img1 = np.array(img1)
plt.imshow(img1,cmap='gray')

img2 = Image.open('Flake000936_Cam2_1_2022-2-13-20-0-36-227.png')
img2 = np.array(img2)
plt.imshow(img2,cmap='gray')
```

```python
ret,thresh1 = cv2.threshold(img1,50,255,cv2.THRESH_BINARY)
plt.imshow(thresh1,cmap='gray')
```
